```
import matplotlib.pyplot as plt
import numpy as np

# Données
labels = [
    'Majuscules', 'Minuscules', 'Chiffres', 'Caractères spéciaux',
    'Maj+Min', 'Maj+Chiffres', 'Maj+Spéciaux', 'Min+Chiffres', 'Min+Spéciaux',
    'Chiffres+Spéciaux', 'Maj+Min+Chiffres', 'Maj+Min+Spéciaux',
    'Maj+Chiffres+Spéciaux', 'Min+Chiffres+Spéciaux', 'Tout'
]

values = [
    6.70, 85.07, 64.89, 0.52, 4.05, 4.83, 0.14, 50.95, 0.48,
    0.34, 3.16, 0.13, 0.12, 0.31, 0.31
]

# Définition des couleurs
colors = plt.cm.viridis(np.linspace(0, 1, len(values)))

# Création d'un camembert optimisé
fig, ax = plt.subplots(figsize=(12, 8))

# Définition de l'explosion pour mieux voir les petites valeurs
explode = [0.1 if v < 1 else 0 for v in values]

# Fonction d'affichage des pourcentages
def autopct_format(pct):
    return ('%1.1f%%' % pct) if pct > 1 else ''

# Création du camembert
wedges, texts, autotexts = ax.pie(values, labels=labels, autopct=autopct_format, startangle=140,
                                   colors=colors, explode=explode, wedgeprops={'edgecolor': 'black'})

# Amélioration des textes
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

# Ajout du titre
ax.set_title("Répartition Des Caractéristiques Des Mots De Passe (Optimisé)", fontsize=14, fontweight='bold')

# Affichage
plt.show()

# Création d'un histogramme pour une meilleure comparaison des valeurs
fig, ax = plt.subplots(figsize=(12, 6))

ax.barh(labels, values, color=colors, edgecolor='black')

ax.set_xlabel('Proportion (%)', fontsize=12)
ax.set_title('Comparaison Des Proportions De Caractéristiques Des Mots De Passe', fontsize=14, fontweight='bold')
ax.grid(axis='x', linestyle='--', alpha=0.7)

# Affichage des valeurs sur les barres
for i, v in enumerate(values):
    ax.text(v + 1, i, f"{v:.1f}%", va='center', fontsize=10, fontweight='bold')

# Affichage
plt.show()

```
